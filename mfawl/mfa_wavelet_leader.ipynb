```python
def _Lj3D(coeffs, level):
    """
    Determines the LOCAL wavelet leader (i.e. corresponding to the combination of filters
    of which the coefficient list is provided in the argument).
    Parameters
    ----------
    coeffs : list of arrays
        The input list of arrays of the coefficients ordered as [cDn, cDn-1,....,cD1], cDn
        is an array with n being the level.
    level : number
        A non-negative integer number. Higher the level, larger the scale and smaller the number of
        coefficients.
        
    Returns
    ------
    out : list
        The list of local wavelet leaders.
        
    Example
    -------
    a,b,c being the wavelet coefficients generated randomly.
    a = np.random.rand(4,4,4)
    b = np.random.rand(8,8,8)
    c = np.random.rand(16,16,16)
    Coeffs_list = [a,b,c]
    Lj = _Lj3D(Coeffs_list,3)
    len(Lj)
    Ans: 8
    """
    if level > len(coeffs) or level < 0:
        raise ValueError("The value of level is out of range [1, {:d}]".format(len(coeffs)))
    L = coeffs[-level:]
    nkx = L[0].shape[0]
    nky = L[0].shape[1]
    nkz = L[0].shape[2]
    nj = len(L)
    Lj = []
    for kx in range(1,nkx-1):
        for ky in range(1,nky-1):
            for kz in range(1,nkz-1):
                neighborhood = []
                for j in range(nj):
                    kxi = 2**j * (kx-1)
                    kxf = 2**j * (kx+2)
                    kyi = 2**j * (ky-1)
                    kyf = 2**j * (ky+2)
                    kzi = 2**j * (kz-1)
                    kzf = 2**j * (kz+2)
                    lambda_jk = L[j][kxi:kxf, kyi:kyf, kzi:kzf]
    #                 print(lambda_jk.shape)
                    flatten_array = lambda_jk.ravel()
                    neighborhood = neighborhood + list(flatten_array)
    #             print(max(neighborhood))
                Lj.append(max(neighborhood))
    return Lj
```

```python
def _keywise_coeffs(levelwise_coeffs):
    """
    This function converts the list of levelwise coefficients into a list of keywise coefficients.
    
    The wavelet coefficients are stored level wise in a list of dictionaries of key-value pair, while
    computing from pywt.wavedecn(). For the computation of wave-leaders, we need to extract the wavelet
    coefficients at each level for a given key. 
    
    Parameter
    --------
    levelwise_coeffs : list of dictionaries
        
    Return
    ------
    out : list of lists
    """
    max_level = len(levelwise_coeffs)
#     print(max_level)
    key_list = list(levelwise_coeffs[1].keys())
#     print(key_list)
    L = []
    for key in key_list:
#         print(key)
        l = []
        for level in range(1,max_level):
            l.append(levelwise_coeffs[level][key])
        L.append(l)
    return L
```
